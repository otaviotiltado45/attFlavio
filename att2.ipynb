```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('aluguel_duas_variaveis_sem_ruido.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['Area'], df['Preco'])
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Área da Casa')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Quartos'], df['Preco'])
plt.xlabel('Número de Quartos')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Número de Quartos')
plt.show()

correlation_matrix = df[['Area', 'Quartos', 'Preco']].corr()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Quartos', y='Preco', data=df)
plt.xlabel('Número de Quartos')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Distribuição de Preços de Aluguel por Número de Quartos')
plt.show()

X_area = df[['Area']]
X_area_quartos = df[['Area', 'Quartos']]
y = df['Preco']

X_train_area, X_test_area, y_train_area, y_test_area = train_test_split(X_area, y, test_size=0.3, random_state=42)
X_train_area_quartos, X_test_area_quartos, y_train_area_quartos, y_test_area_quartos = train_test_split(X_area_quartos, y, test_size=0.3, random_state=42)

model_area = LinearRegression()
model_area.fit(X_train_area, y_train_area)
y_pred_area = model_area.predict(X_test_area)
r2_area = r2_score(y_test_area, y_pred_area)
rmse_area = mean_squared_error(y_test_area, y_pred_area, squared=False)

model_area_quartos = LinearRegression()
model_area_quartos.fit(X_train_area_quartos, y_train_area_quartos)
y_pred_area_quartos = model_area_quartos.predict(X_test_area_quartos)
r2_area_quartos = r2_score(y_test_area_quartos, y_pred_area_quartos)
rmse_area_quartos = mean_squared_error(y_test_area_quartos, y_pred_area_quartos, squared=False)

print("\nModelo de Regressão Linear com apenas uma variável (Área):")
print("R²:", r2_area)
print("RMSE:", rmse_area)

print("\nModelo de Regressão Linear com duas variáveis (Área e Quartos):")
print("R²:", r2_area_quartos)
print("RMSE:", rmse_area_quartos)
```
